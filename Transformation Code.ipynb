# Data Modelling

### Now we have to convert this flat data into the fact table and dimension data and this process is generally called as dimensional modelling 

### This is generally done to divided the information or column of data which will remain static for long time and data which remain changing after some particular time so that we can easily see the final results by seeing the fact table information or columns , this is generally done when we have lot of information or columns in data 

### We will do this using one online tool available which is called as lucid chart

### This time we have only one table and less number of columns but when you are working with real time data then you will have many number of tables with lot amount of columns then you have to put much time to understand how to build fact and dimension table of these data

- so let's start with dimensional modelling by understanding what will be our dimension tables and what will be out fact tables for the sake of this table , we will be creating mutiple dimension table even the information or columns that would have been inside the fact table , we will put that into dimension tables so we can understand how to work with any real world table

- remeber , always add first row as the primay key or pk and with this should have the corresponding second column with either the name of information or column present in data or make the "id" column of that particular fact table or dimension table

# Transformation Code

```python
# Now we convert our flat file into the fact and dimesion table according the above data modelling 

# first do df.info() to check that any column of data should be not have datatype as object except the column which store text or string otherwise change the datatype to int or float64 and for columns of date there dataype should be datatime other wise change it into the date time
df.info()
```

```python
# as date related columns tpep_pickup_datetime and tpep_dropoff_datetime have datatype as object so we change it's datatype to datetime
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
# this all is done so that we can extract year,month,day easily
```

```python
df.info()
```

```python
# dropping the duplicates
df = df.drop_duplicates()
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
# Now we make one by one df for each table (dimension table with same columns as data in second column of dimension table)
# and the data of first column of dimension table will be used when we are applying the sql quary on the data
datetime_dim = df[['tpep_pickup_datetime','tpep_dropoff_datetime']].reset_index(drop=True)

datetime_dim['tpep_pickup_datetime'] = datetime_dim['tpep_pickup_datetime']
datetime_dim['pick_hour'] = datetime_dim['tpep_pickup_datetime'].dt.hour
datetime_dim['pick_day'] = datetime_dim['tpep_pickup_datetime'].dt.day
datetime_dim['pick_month'] = datetime_dim['tpep_pickup_datetime'].dt.month
datetime_dim['pick_year'] = datetime_dim['tpep_pickup_datetime'].dt.year
datetime_dim['pick_weekday'] = datetime_dim['tpep_pickup_datetime'].dt.weekday

datetime_dim['tpep_dropoff_datetime'] = datetime_dim['tpep_dropoff_datetime']
datetime_dim['drop_hour'] = datetime_dim['tpep_dropoff_datetime'].dt.hour
datetime_dim['drop_day'] = datetime_dim['tpep_dropoff_datetime'].dt.day
datetime_dim['drop_month'] = datetime_dim['tpep_dropoff_datetime'].dt.month
datetime_dim['drop_year'] = datetime_dim['tpep_dropoff_datetime'].dt.year
datetime_dim['drop_weekday'] = datetime_dim['tpep_dropoff_datetime'].dt.weekday

# make one column as "datetime_id"  as index of df as "datetime_id" is primary key means should be unique and in df index value is the unique to each row 
datetime_dim['datetime_id'] = datetime_dim.index
```

```python
# to get the df with particular order of columns
datetime_dim = datetime_dim[['datetime_id', 'tpep_pickup_datetime', 'pick_hour', 'pick_day', 'pick_month', 'pick_year', 'pick_weekday',
                             'tpep_dropoff_datetime', 'drop_hour', 'drop_day', 'drop_month', 'drop_year', 'drop_weekday']]
```

```python
datetime_dim.head()
```

```python
passenger_count_dim = df[['passenger_count']].reset_index(drop=True)
passenger_count_dim['passenger_count_id'] = passenger_count_dim.index
passenger_count_dim = passenger_count_dim[['passenger_count_id','passenger_count']]
```

```python
passenger_count_dim
```

```python
trip_distance_dim = df[['trip_distance']].reset_index(drop=True)
trip_distance_dim['trip_distance_id'] = trip_distance_dim.index
trip_distance_dim = trip_distance_dim[['trip_distance_id','trip_distance']]
```

```python
trip_distance_dim
```

```python
rate_code_type = {
    1:"Standard rate",
    2:"JFK",
    3:"Newark",
    4:"Nassau or Westchester",
    5:"Negotiated fare",
    6:"Group ride"
}

rate_code_dim = df[['RatecodeID']].reset_index(drop=True)
rate_code_dim['rate_code_id'] = rate_code_dim.index
rate_code_dim['rate_code_name'] = rate_code_dim['RatecodeID'].map(rate_code_type)
rate_code_dim = rate_code_dim[['rate_code_id','RatecodeID','rate_code_name']]
```

```python
rate_code_dim.head()
```

```python
pickup_location_dim = df[['pickup_longitude', 'pickup_latitude']].reset_index(drop=True)
pickup_location_dim['pickup_location_id'] = pickup_location_dim.index
pickup_location_dim = pickup_location_dim[['pickup_location_id','pickup_latitude','pickup_longitude']]
```

```python
pickup_location_dim.head()
```

```python
dropoff_location_dim = df[['dropoff_longitude', 'dropoff_latitude']].reset_index(drop=True)
dropoff_location_dim['dropoff_location_id'] = dropoff_location_dim.index
dropoff_location_dim = dropoff_location_dim[['dropoff_location_id','dropoff_latitude','dropoff_longitude']]
```

```python
dropoff_location_dim.head()
```

```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}
payment_type_dim = df[['payment_type']].reset_index(drop=True)
payment_type_dim['payment_type_id'] = payment_type_dim.index
payment_type_dim['payment_type_name'] = payment_type_dim['payment_type'].map(payment_type_name)
payment_type_dim = payment_type_dim[['payment_type_id','payment_type','payment_type_name']]
```

```python
payment_type_dim.head()
```

```python
# Create the fact table 
# to make the fact table we are merging all the df of dimesion table with original df with certain column

# example
# Merge DataFrames based on different key column names
# merged_df = pd.merge(df1, df2, left_on='key_left', right_on='key_right', how='inner')

# DataFrame 1:
#   key_left  value_df1
# 0        A          1
# 1        B          2
# 2        C          3
# 3        D          4

# DataFrame 2:
#   key_right  value_df2
# 0         B          5
# 1         D          6
# 2         E          7
# 3         F          8

# Merged DataFrame:
#   key_left  value_df1 key_right  value_df2
# 0        B          2         B          5
# 1        D          4         D          6
```

```python
# naming index column of df dataframe with name "trip_id"
df['trip_id'] = df.index
```

```python
fact= df.merge(passenger_count_dim, left_on='trip_id', right_on='passenger_count_id')
```

```python
fact
```

```python
# make df with some number of columns
fact=fact[['trip_id','VendorID','passenger_count_id']]
```

```python
fact
```

```python
fact_table = df.merge(passenger_count_dim, left_on='trip_id', right_on='passenger_count_id') \
             .merge(trip_distance_dim, left_on='trip_id', right_on='trip_distance_id') \
             .merge(rate_code_dim, left_on='trip_id', right_on='rate_code_id') \
             .merge(pickup_location_dim, left_on='trip_id', right_on='pickup_location_id') \
             .merge(dropoff_location_dim, left_on='trip_id', right_on='dropoff_location_id')\
             .merge(datetime_dim, left_on='trip_id', right_on='datetime_id') \
             .merge(payment_type_dim, left_on='trip_id', right_on='payment_type_id') \
             [['trip_id','VendorID', 'datetime_id', 'passenger_count_id',
               'trip_distance_id', 'rate_code_id', 'store_and_fwd_flag', 'pickup_location_id', 'dropoff_location_id',
               'payment_type_id', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount',
               'improvement_surcharge', 'total_amount']]
```
